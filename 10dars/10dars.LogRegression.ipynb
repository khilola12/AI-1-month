```python
# 1.Data preprocessinig

import pandas as pd
df=pd.read_csv('movie_genre_classification_final.csv')
```

```python
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
# - Missing Values toldirish


for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.info()
```

```python
df.nunique()
```

```python
# Encoding

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for ustun in df.columns:
    if df[ustun].dtype=='object':
        if df[ustun].nunique()<=4:
            dummies=pd.get_dummies(df[ustun],prefix=ustun,dtype=int)
            df=pd.concat([df.drop(columns=ustun),dummies])
        else:
            df[ustun]=encoder.fit_transform(df[ustun])
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

for ustun in df.columns:
    if df[ustun].dtype!='object'and ustun != 'Genre':
        df[ustun]=scaler.fit_transform(df[[ustun]])
```

```python
# x,y Input Output larni aniqlash:

x=df.drop('Genre', axis=1)

y=df['Genre']
x=df.drop('Genre', axis=1)
```
