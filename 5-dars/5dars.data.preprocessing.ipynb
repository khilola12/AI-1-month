```python
import pandas as pd
```

```python
df=pd.read_csv('test.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Classlar sonini aniqlash
df['Race'].nunique()
```

```python
#1-bosqich
#Soxta(dummies) ustunlar yaratish
#
#pandas kutubxonasidan dummies funksiyasini chaqirish -> (pd.get_dummies) ,
#qaysi ustunni dummies qilib yasash -> (df['Race']) ,
#ustunga nom berish -> (prefix='ustuncha') ,
#data type ni kiritish -> (dtype=int).
dummies=pd.get_dummies(df['Race'],prefix='ustuncha',dtype=int)
```

```python
dummies
```

```python
df.info()
```

```python
#2-bosqich
#Datasetni birlashtirish (bunda eski ustun tashlab yuboriladi)
df=pd.concat([df.drop(columns='Race'),dummies],axis=1)
```

```python
df.nunique()
```

```python
#1-bosqich
#Soxta(dummies) ustunlar yaratish
dummies=pd.get_dummies(df['SchoolType'],prefix='columncha',dtype=int)
```

```python
dummies
```

```python
#2-bosqich
#Datasetni birlashtirish --> pd.concat  (bunda eski ustun tashlab yuboriladi)
df=pd.concat([df.drop(columns='SchoolType'),dummies],axis=1)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
dummies=pd.get_dummies(df['Gender'],prefix='columncha',dtype=int)
```

```python
df=pd.concat([df.drop(columns='Gender'),dummies],axis=1)
```

```python
df.info()  #['Race'], ['SchoolType'] va ['Gender'] ustunlari (dummies yaratilib) tashlab yuborildi.
```

```python
#Kutubxonani chaqirish
#scikit-learn kutubxonasidan (from sklearn.)   -> preprocessing xonasidan -> LabelEncoder ni import qilish:
from sklearn.preprocessing import LabelEncoder
```

```python
encoder
```

```python
df.info()
```

```python
df.nunique()
```

```python
# object dan  numerical ga o'tkazish -->  fit_transform
df['Locale']=encoder.fit_transform(df['Locale'])
```

```python
df['ParentalEducation']=encoder.fit_transform(df['ParentalEducation'])
```
